```python
current = '.'
speech_path = path.join(current, 'hw3_speech_files')
```

```python
files = listdir(speech_path)
```

```python
def getIntensity(sound_file):
    Intensity = call(sound_file, "To Intensity", 100, 0.0, 600)
    minIntensity = call(Intensity, "Get minimum", 0.0, 0.0, "Parabolic")
    maxIntensity = call(Intensity, "Get maximum", 0.0, 0.0, "Parabolic")
    meanIntensity = call(Intensity, "Get mean", 0.0, 0.0, "energy")
    Minintensity.append(minIntensity)
    Maxintensity.append(maxIntensity)
    Meanintensity.append(meanIntensity)
    
    
```

```python
def getPitch(sound_file):
    Pitch = call(sound_file, "To Pitch (ac)", 0.0, 75, 15, 'off', 0.03, 0.45, 0.01, 0.35, 0.14, 600)
    minPitch = call(Pitch, "Get minimum", 0.0, 0.0, "Hertz", "Parabolic")
    maxPitch = call(Pitch, "Get maximum", 0.0, 0.0, "Hertz", "Parabolic")
    meanPitch = call(Pitch, "Get mean", 0.0, 0.0, "Hertz")
    Minpitch.append(minPitch)
    Maxpitch.append(maxPitch)
    Meanpitch.append(meanPitch)
    
    
```

```python
def stripFile(wave_file):
    parts = wave_file.split("_")
    speaker = parts[0]
    session = parts[1]
    emotion = parts[2]
    start_time = float(parts[3])
    rem = parts[-1].split('.')[0]
    words = rem.replace('-', ' ')
    
    return speaker, session, emotion, start_time, words
```

```python
Emotion = []
Minpitch = []
Maxpitch = []
Meanpitch = []
Minintensity = []
Maxintensity = []
Meanintensity = []
Speaker = []
Session = []
Start_Time = []
Words = []
```

```python
def processFiles(wave_file):
    base_dir = 'hw3_speech_files/'
    print("Processing {}...".format(wave_file))
    speaker, session, emotion, start_time, words = stripFile(wave_file)
    Speaker.append(speaker)
    Session.append(session)
    Start_Time.append(start_time)
    Emotion.append(emotion)
    Words.append(words)
    sound_file = parselmouth.Sound(base_dir + wave_file)
    getIntensity(sound_file)
    getPitch(sound_file)
```

```python
for file in files:
    processFiles(file)
```

```python
dictionary = {
    'Speaker' : Speaker,
    'Session' : Session,
    'Start Time' : Start_Time,
    'Words' : Words,
'Min Pitch': Minpitch,
'Max Pitch': Maxpitch,
'Mean Pitch': Meanpitch,
'Min Intensity': Minintensity,
'Max Intensity': Maxintensity,
'Mean Intensity': Meanintensity,
'Target': Emotion
}
dataframe = pd.DataFrame(data=dictionary)
dataframe.head()
```
